```python
from zipfile import ZipFile
import os

base = "/mnt/data/ntag424-personalizer"

structure = {
    "notebooks": [
        "01_FindReader.ipynb",
        "02_ReadTag.ipynb",
        "03_Authenticate.ipynb",
        "04_EnableSUN.ipynb",
        "05_Personalize.ipynb",
    ],
    "personalizer": [
        "__init__.py",
        "reader.py",
        "apdu.py",
        "ntag424.py",
        "crypto.py",
        "sun.py",
    ],
    "exports": [".gitkeep"],
    "keys": [".gitkeep"],
    "data": [".gitkeep"],
}

os.makedirs(base, exist_ok=True)

readme = """# NTAG424 Personalizer

Project scaffold for an NTAG424 DNA personalization toolkit.

## Folders

- notebooks/ - Jupyter notebooks
- personalizer/ - Python modules
- data/ - Input CSVs and metadata
- keys/ - Local key storage (do not commit secrets)
- exports/ - Generated outputs

"""

with open(os.path.join(base, "README.md"), "w", encoding="utf-8") as f:
    f.write(readme)

gitignore = """__pycache__/
*.pyc
.ipynb_checkpoints/
.venv/
keys/*
exports/*
!.gitkeep
"""

with open(os.path.join(base, ".gitignore"), "w", encoding="utf-8") as f:
    f.write(gitignore)

for folder, files in structure.items():
    os.makedirs(os.path.join(base, folder), exist_ok=True)
    for fn in files:
        path = os.path.join(base, folder, fn)
        if fn.endswith(".ipynb"):
            content = {
                "cells": [],
                "metadata": {
                    "kernelspec": {
                        "display_name": "Python 3",
                        "language": "python",
                        "name": "python3"
                    },
                    "language_info": {"name": "python"}
                },
                "nbformat": 4,
                "nbformat_minor": 5
            }
            import json
            with open(path, "w", encoding="utf-8") as f:
                json.dump(content, f, indent=2)
        elif fn.endswith(".py"):
            with open(path, "w", encoding="utf-8") as f:
                if fn == "__init__.py":
                    f.write('"""NTAG424 Personalizer package."""\n')
                else:
                    mod = os.path.splitext(fn)[0]
                    f.write(f'"""Module: {mod}"""\n\n')
            # .gitkeep
        else:
            open(path, "a").close()

zip_path = "/mnt/data/ntag424-personalizer.zip"
with ZipFile(zip_path, "w") as z:
    for root, _, files in os.walk(base):
        for file in files:
            full = os.path.join(root, file)
            arc = os.path.relpath(full, os.path.dirname(base))
            z.write(full, arc)

print(zip_path)
```
